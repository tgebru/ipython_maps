```python
#Map NYC income, race, education district level predictions
#Map District level predictions for NYC race, education, income
import json 
import matplotlib.pyplot as plt
import numpy as np
import random
import os
%matplotlib inline
plt.rcParams['figure.figsize']=(20,20)
```

```python
#Specify a colormap 
#1. Bin values by 20%
#2. Rank districts from 1:10
```

```python
#Draw NYC district boundary files
#Load NYC district files
boundary_fname='/imagenetdb3/tgebru/cvpr2016/housing_shapefiles/\
nyc-district-tabulation-areas-polygons.geojson'

years=range(2007,2015)
x_list=[]
y_list=[]

fig = plt.figure() 
ax = fig.gca()
    
for year in years:
    with open(boundary_fname) as json_file:
        json_data=json.load(json_file)

    for feat in json_data['features']:
        coords = feat['geometry']['coordinates'] 
        for c in coords:
            x_list.append([i for i,j in c[0]])
            y_list.append([j for i,j in c[0]])
    
    for x,y in zip(x_list,y_list):
        ax.plot(x,y,'k',linewidth=2.0)
```
